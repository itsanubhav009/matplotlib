```python
#linear regression using single variable

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Book 2(Sheet1).csv")
df

df.columns = df.columns.str.strip()
#for removeing any unexpected space or anything
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[1500]])  
```

```python
reg.coef_  #m
```

```python
reg.intercept_   #b
```

```python
126.67682927*5000 +  213033.53658536595
```

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_csv("prediction.csv" , index = False)
```
